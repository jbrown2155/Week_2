```python
# Import libraries and dependencies
import pandas as pd
from pathlib import Path
import numpy as np
```

```python
# Set the path
file_path = Path('PyBank_Data.csv')

# Read in the CSV as a DataFrame
financial_analysis = pd.read_csv(file_path)
financial_analysis.head()
```

```python
financial_analysis["shifted_column"] = financial_analysis["Profit/Losses"].shift(1)
financial_analysis["difference"] = financial_analysis["Profit/Losses"] - financial_analysis["shifted_column"]
financial_analysis
```

```python
financial_analysis["Profit/Losses"]=financial_analysis["Profit/Losses"].astype("float")
financial_analysis["shifted_column"]=financial_analysis["shifted_column"].astype("float")
financial_analysis["difference"]=financial_analysis["difference"].astype("float")
financial_analysis
```

```python
pd.options.display.float_format = "${:,.2f}".format
```

```python
financial_analysis
```

```python
Total_months = financial_analysis["Date"].count() 
Total = '{:.0f}'.format(financial_analysis["Profit/Losses"].sum())
Mean = '{:.2f}'.format(financial_analysis["difference"].mean())
Best = '(${:.0f})'.format(financial_analysis["difference"].max())
Worst = '(${:.0f})'.format(financial_analysis["difference"].min())
best_month=financial_analysis.sort_values("difference"), ("date")
worst_month=financial_analysis.sort_values("difference", ascending=False), ("date")
```

```python
# Print Required Statistics
print("Financial Analysis")
print("_____________________________")
print(f"Total Months: {Total_months}")
print(f"Total: ${Total}")
print(f"Average Change: ${Mean}")
print(f"Greatest Increase in Profits: {Best}")
print(f"Greatest Decrease in Profits: {Worst}")                   
```
